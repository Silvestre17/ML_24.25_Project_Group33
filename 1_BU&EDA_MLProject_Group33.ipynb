The New York Workersâ€™ Compensation Board (WCB) administers and regulates workersâ€™ compensation, disability, volunteer firefighters, volunteer ambulance workers, and volunteer civil defence  workersâ€™ benefits. As the regulating authority, the WCB is  responsible  for  **assembling and deciding on claims whenever it becomes aware of a workplace injury**. Since 2000, the WCB has assembled and reviewed more than 5 million claims.

However, manually reviewing all claims is an arduous and time-consuming process. For that reason, the WCB has reached out to Nova IMS to assist them in the creation of a model that can automate the decision-making whenever a new claim is received.

```python
# Create a new column 'Birth Year' and 'Age at Injury' for clean
train_data['Birth Year Clean'] = train_data['Birth Year']
train_data['Age at Injury Clean'] = train_data['Age at Injury']

test_data['Birth Year Clean'] = test_data['Birth Year']
test_data['Age at Injury Clean'] = test_data['Age at Injury']
```

```python
# Check the cases where the age is 0 and the birth year is complete
train_data[(train_data['Age at Injury'] == 0) & (train_data['Birth Year'].notnull())]
```

```python
# Replace 'Birth Year' of 0 with birth year calculated from 'Age at Injury' and 'Assembly Date' only if the age is different from 0
# - Do it in the '...Clean' column
train_data.loc[(train_data['Birth Year'] == 0) & (train_data['Age at Injury'] != 0), 'Birth Year Clean'] = train_data['Assembly Date'].dt.year - train_data['Age at Injury']
test_data.loc[(test_data['Birth Year'] == 0) & (test_data['Age at Injury'] != 0), 'Birth Year Clean'] = test_data['Assembly Date'].dt.year - test_data['Age at Injury']

# Check if there are still cases where the birth year is 0
print('The number of cases where the birth year is 0 [Train]:', train_data[train_data['Birth Year Clean'] == 0].shape[0], '(', round(train_data[train_data['Birth Year Clean'] == 0].shape[0] / train_data.shape[0] * 100, 2), '%)')
print('The number of cases where the birth year is 0 [Test]:', test_data[test_data['Birth Year Clean'] == 0].shape[0], '(', round(test_data[test_data['Birth Year Clean'] == 0].shape[0] / test_data.shape[0] * 100, 2), '%)')
```

```python
# Impute missing values in the Age at Injury column
# Replace 'Age at Injury' values with calculated age only where the age is 0 and the birth year is complete and different from 0
# - Do it in the '...Clean' column
train_data.loc[(train_data['Age at Injury'] == 0) & (train_data['Birth Year'] != 0), 'Age at Injury Clean'] = train_data['Assembly Date'].dt.year - train_data['Birth Year']
test_data.loc[(test_data['Age at Injury'] == 0) & (test_data['Birth Year'] != 0), 'Age at Injury Clean'] = test_data['Assembly Date'].dt.year - test_data['Birth Year']

# Check if there are still cases where the age is 0
print('The number of cases where the age is 0 [Train]:', train_data[train_data['Age at Injury Clean'] == 0].shape[0], '(', round(train_data[train_data['Age at Injury Clean'] == 0].shape[0] / train_data.shape[0] * 100, 2), '%)')
print('The number of cases where the age is 0 [Test]:', test_data[test_data['Age at Injury Clean'] == 0].shape[0], '(', round(test_data[test_data['Age at Injury Clean'] == 0].shape[0] / test_data.shape[0] * 100, 2), '%)')
```

```python
# Replace the rest of the 0 values in the 'Age at Injury' and 'Birth Year' with NaN
train_data.loc[train_data['Age at Injury Clean'] == 0, 'Age at Injury Clean'] = np.nan
train_data.loc[train_data['Birth Year Clean'] == 0, 'Birth Year Clean'] = np.nan

test_data.loc[test_data['Age at Injury Clean'] == 0, 'Age at Injury Clean'] = np.nan
test_data.loc[test_data['Birth Year Clean'] == 0, 'Birth Year Clean'] = np.nan
```

```python
# Check the distribution of 'Age at Injury' and 'Birth Year' after the imputation
fig, ax = plt.subplots(1, 2, figsize=(20, 6))

# Plotting 'Age at Injury'
sns.histplot(train_data['Age at Injury'], kde=True, color='#FFB800', ax=ax[0])
ax[0].set_title('Age at Injury', fontsize=16, fontweight='bold')
ax[0].set_xlabel('Age', fontsize=12, fontweight='bold')
ax[0].set_ylabel('Frequency', fontsize=12, fontweight='bold')
    
# Plotting 'Birth Year'
sns.histplot(train_data['Birth Year'], kde=True, color='#CC9300', ax=ax[1])
ax[1].set_title('Birth Year', fontsize=16, fontweight='bold')
ax[1].set_xlabel('Year', fontsize=12, fontweight='bold')
ax[1].set_ylabel('Frequency', fontsize=12, fontweight='bold')

sns.despine(top=True, right=True)
plt.tight_layout()
plt.show()
```

> The distribution of **`Age at Injury`** and **`Birth Year`** seems more consistent after the imputation.
